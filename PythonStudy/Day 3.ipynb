"이런 작업에 얼만큼의 코드가 실제로 작성되어야 하나요?"

```python
def 기상청_연도별_정리(excel_in, excel_out, 연도목록):
    날씨 = office.Excel(excel_in)
    # 수원날씨 선택
    수원날씨 = 날씨[날씨['location']=='suwon']
    # 일부 열 선택
    수원날씨 = 수원날씨[['date', 'avgTemp', 'weather']]
    # 열제목 변경
    열제목 = {'date': '날짜', 'weather': '기상', 'avgTemp': '일평균'}
    수원날씨 = 수원날씨.rename(columns=열제목)

    # 연도별 선택
    수원날씨.set_index('날짜')
    
    표목록 = []
    for 연도 in 연도목록:
        표 = 수원날씨.ix[str(연도)]
        
    # 연도별 기록을 각 시트로 저장
    office.to_excel(표목록, excel_out)
```

```python
기상청_연도별_정리('data/weather.xlsx', '수원 연도별.xlsx', [2013, 2014, 2015])
```

```python
html
```

```python
파일 = open('네이버 금융 링크.csv', 'w', encoding='cp949')

for 태그 in html.find_all('a'):
    url = 태그.get('href', '')
    제목 = 태그.text
    if url.startswith('http'):        
        항목 = '{0},{1}\n'.format(제목, url)
        파일.write(항목)
        
파일.close()
```

! 웹 스크래핑에서 아주 유의하셔야 하는 점이 있습니다.

```python
url = 'http://news.jtbc.joins.com/'
```

```python
web.download(url, 'jtbc_news_home.html')
```

```python
html = web.parse_html('jtbc_news_home.html')
```

```python
html.title
```

```python
파일 = open('jtbc-2017-05-30T2000.csv', 'w', encoding='cp949')

for 태그 in html.find_all('a'):
    제목 = 태그.text
    if '대통령' in 제목:
        url = 태그.get('href', '')
        #print(제목, url)
        항목 = '{0}\t{1}\n'.format(제목, url)
        파일.write(항목)
        
파일.close()
```
